```python
import pandas as pd
datos = pd.read_excel ("/content/drive/MyDrive/ESTADISTICA INDUSTRIAL/Corte 3/datos_regresion_simple.xlsx")
datos.head()
```

```python
# prompt: Necesito una grafica de dispersion  interactivo con colores verde y morado con linea de tendencia con x=Horas_Trabajadas y y=Productos_Terminados

import plotly.express as px

fig = px.scatter(datos, x="Horas_Trabajadas", y="Productos_Terminados", trendline="ols",
                 color_discrete_sequence=["green", "purple"])
fig.show()
```

```python
# prompt: Dame la ecuacion de la recta de la grafica

import statsmodels.api as sm

# Definir las variables dependiente (y) e independiente (x)
y = datos['Productos_Terminados']
x = datos['Horas_Trabajadas']

# Añadir una constante al modelo para el intercepto
x = sm.add_constant(x)

# Crear y ajustar el modelo de regresión lineal
modelo = sm.OLS(y, x).fit()

# Obtener el resumen del modelo
print(modelo.summary())

# La ecuación de la recta es y = intercepto + pendiente * x
# El intercepto y la pendiente se encuentran en el resumen del modelo
# Por ejemplo, si el resumen indica:
# const      10.5000  (Intercepto)
# Horas_Trabajadas  2.0000  (Pendiente)
# La ecuación sería: Productos_Terminados = 10.50 + 2.00 * Horas_Trabajadas

# Puedes acceder a los coeficientes directamente:
intercepto = modelo.params['const']
pendiente = modelo.params['Horas_Trabajadas']

print(f"\nLa ecuación de la recta de regresión es: Productos_Terminados = {intercepto:.2f} + {pendiente:.2f} * Horas_Trabajadas")
```
